```python
import pandas as pd

produtos = {'registro_produto': ['laptop', 'impressora', 'tablet', 'mesa', 'cadeira'],
        'preço': [1200, 150, 300, 450, 200]}

df_produtos = pd.DataFrame(produtos, index=['produto_1', 'produto_2', 'produto_3', 'produto_4', 'produto_5'])
df_produtos
```

```python
vendas = {'registro_produto': ['laptop', 'tablet', 'laptop', 'mesa', 'cadeira', 'mesa', 'tablet', 'cadeira', 'cadeira', 'laptop', 'tablet', 'cadeira'],
          'loja': ['01', '02', '01', '01', '03', '02', '01', '02', '03', '02', '02', '03'],
          'vendedor': ['Ana', 'João', 'Cláudia', 'Maria', 'Maria', 'João', 'João', 'Maria', 'Maria', 'Maria', 'Maria', 'Maria'],
        'data': ['03/05/2023', '17/11/2022', '09/08/2022', '14/05/2023', '23/07/2021', '09/09/2021', '19/08/2023', '23/07/2021', '23/07/2021', '09/08/2022', '24/01/2021', '18/03/2023']}


df_vendas = pd.DataFrame(vendas, index=['venda_1', 'venda_2', 'venda_3', 'venda_4', 'venda_5', 'venda_6', 'venda_7', 'venda_8', 'venda_9', 'venda_10', 'venda_11', 'venda_12'])
df_vendas
```

### Groupby

```python
# Get Group 'Maria'
df_vendas.groupby('vendedor').get_group('Maria')
```

```python
total = pd.DataFrame.merge(df_vendas,df_produtos,on='registro_produto')
total
```

```python
total.groupby('registro_produto')[['preço']].sum()
```

```python
# Ordena e pega a primeira linha de cada grupo
total.sort_values(by='preço',    
               ascending=True)
```

```python
teste = total.sort_values(by='preço', ascending=True).groupby('vendedor')#.first()
for key, item in teste:
    print(teste.get_group(key), "\n\n")
```

```python
total.sort_values(by='preço', ascending=True).groupby('vendedor').first()
```

```python
# Agregando
total.groupby('vendedor').agg({'loja':pd.Series.nunique, 
                           'preço':'sum',
                           'registro_produto':pd.Series.mode})
```

### Pivot table

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_openml

data,target = fetch_openml("autos", version=1, as_frame=True, return_X_y=True)
data
```

```python
data['fuel-type'].unique()
```

<img src="https://builtin.com/sites/www.builtin.com/files/styles/ckeditor_optimize/public/inline-images/1_pandas-pivot-table.jpeg" width=400 />

```python
pivot = np.round(pd.pivot_table(data, values='price', # o dado que quero mostrar 
                                index='num-of-doors', 
                                columns='fuel-type', 
                                aggfunc=np.mean),2)
pivot
```

```python
pivot = np.round(pd.pivot_table(data, values='price', 
                                index=['num-of-doors', 'body-style'], 
                                columns=['fuel-type', 'fuel-system'], 
                                aggfunc=np.mean,
                                fill_value=0),2)
pivot
```
